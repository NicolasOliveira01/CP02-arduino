```python
import os
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
modelo = load_model('modelo_mnist.h5')
# carrega o modelo do mnist_2
```

```python
# Definir o caminho da imagem
caminho_imagem = os.path.join("Digito Teste", "sete.jpg")
# os.path.join -> montar caminhos de arquivos, seleciona a pasta "Digito Teste" e o arquivo "sete.jpg"

imagem = load_img(caminho_imagem)
# load_img -> carrega a imagem 

# Exibir a imagem
plt.figure(figsize=(4, 4))
plt.imshow(imagem)
plt.title("Imagem Importada")
plt.axis("off")  # Ocultar os eixos
plt.show()
```

```python
print(f"Modo da imagem: {imagem.mode}")
print(f"Tamanho da imagem: {imagem.size}")
```

```python
imagem_cinza = imagem.convert("L")  # Converte para modo 'L'
# .convert("L") -> converte a imagem para a escala cinza, onde cada pixel é 0 (branco) a 225 (preto) e a imagem passa a ter
# apenas um canal 

imagem_cinza = imagem_cinza.resize((28, 28))
# resize((28, 28)) -> deixa a imagem com 28x28 pixels, o tamanho padrão do MNIST

imagem_array = img_to_array(imagem_cinza)
# img_to_array -> converte a imagem PIL para um array NumPy (28, 28, 1)

print(f"Formato da imagem: {imagem_array.shape}")

# Exibir a imagem
plt.figure(figsize=(4, 4))
plt.imshow(imagem_array)
plt.title("Imagem Importada")
plt.axis("off")  # Ocultar os eixos
plt.show()

# Imprimir os valores numéricos da matriz da imagem
print("Matriz da Imagem (valores numéricos):")
print(imagem_array.squeeze())
# remove todas as dimensões de tamanho 1 do array, ou seja, transforma (28, 28, 1) em (28, 28)
```

```python
imagem_array /= 255.0  
# divide todos os pixels por 255, normalizando para [0, 1]

# Exibir a imagem
plt.figure(figsize=(4, 4))
plt.imshow(imagem_array)
plt.title("Imagem Importada")
plt.axis("off")  # Ocultar os eixos
plt.show()

# Imprimir os valores numéricos da matriz da imagem
print("Matriz da Imagem (valores numéricos):")
print(imagem_array.squeeze())
```

```python
# Exibir a imagem em escala de cinza
plt.figure(figsize=(4, 4))
plt.imshow(imagem_array.squeeze(), cmap="gray")  # Usar cmap="gray" para escala de cinza
plt.title("Imagem em Escala de Cinza")
plt.axis("off")  # Ocultar os eixos
plt.show()

# Imprimir os valores numéricos da matriz da imagem
print("Matriz da Imagem (valores numéricos):")
print(imagem_array.squeeze())
```

```python
# Adicionar uma dimensão para representar o batch
imagem_array = imagem_array.reshape(1, 28, 28, 1)

# Fazer a predição
predicao = modelo.predict(imagem_array)

# Obter o rótulo predito
label_predito = np.argmax(predicao)

# Exibir os resultados
print(f"Predição (valores das probabilidades): {predicao}")
print(f"Rótulo predito: {label_predito}")
```
